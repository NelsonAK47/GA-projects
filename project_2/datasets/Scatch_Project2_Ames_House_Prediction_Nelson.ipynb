```python
#Get all categorical plots

#To plot count plots for all categorical features (To sieve out categories with very uneven high counts that can cause bias)

def countplotter(df, y, row , ax):
    sns.countplot(
        data=df,
        y=y,
        order = df.groupby(y)[['SalePrice']].mean().sort_values(by='SalePrice').index,
        ax=ax,
        palette = 'rainbow',
        edgecolor=sns.color_palette("dark:black", 3))
    threshold_line = ax.axvline(0.7 * row, ls='--', c='red', label='Threshold Line')
    
#To plot boxplot for all categorical features (To check on the distribution in general)
def boxplotter(df, x, y, ax):
        sns.boxplot(data=df,
                    x=x,
                    y=y,
                    ax=ax, palette = 'rainbow',
                    order = df.groupby(y)[[x]].mean().sort_values(by=x).index,
                    orient ='h')
        ax.set_xlabel(x)
        ax.yaxis.label.set_size(12)
        ax.xaxis.label.set_size(12)
        
#Fit both plots at one-go
def categorical_plotter(df, categorical_ft):
    for col in categorical_ft:
        number_of_cat = len(df[col].unique())
        graph_height = int(number_of_cat/2)
        fig, ax = plt.subplots(nrows=1,
                               ncols=2,
                               figsize=(14, graph_height),
                               sharey=True,
                              gridspec_kw={'width_ratios': [1,3]} 
                              );
        ax=ax.ravel();
        row = len(df)
        countplotter(df, col, row, ax[0]);
        boxplotter(df,'SalePrice',col, ax[1]);
        plt.rcParams.update({'figure.max_open_warning': 0})
```

```python
#Get all numerical plots
def scatterplotter(df, columns, y_target):
    ncols = 4
    nrows = int((len(columns)/ncols))
    
    fig, ax = plt.subplots(nrows=nrows, ncols=ncols, figsize=(16, nrows*4))
    ax = ax.ravel()
    for i, col in enumerate(columns):
        sns.regplot(data=df,
                    x=col,
                    y=y_target,
                    ax=ax[i],
                    line_kws={'color':'brown','lw':5})
        ax[i].set_xlabel('')
        ax[i].yaxis.label.set_size(12)
        ax[i].set_title(col, fontsize=10);
                
    plt.tight_layout()
    plt.suptitle('Scatterplot of Numerical Features vs Sale Price', fontsize=30, y = 1.01)
```
